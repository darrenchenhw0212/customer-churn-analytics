## 1. Project Overview

- data exploration and quality assessment
- feature engineering and preprocessing
- model development and evaluation
- explainable AI using SHAP and LIME

Among five supervised machine learning algorithms evaluated, **Gradient Boosting** consistently achieved the strongest predictive performance and was selected as the champion model.

### Key Findings

The modelling and explainability analyses consistently produced the following findings:

### Interpretation

Customers displaying declining engagement should be identified early and offered personalised interventions before churn occurs.

## 4. Deployment Considerations

Before deployment in a production banking environment, several operational considerations should be addressed.

### Prediction Frequency

The model should be executed regularly (for example, weekly or monthly) to continuously identify customers whose behavioural patterns indicate elevated churn risk.

### Model Monitoring

Model performance should be monitored over time to detect concept drift arising from changing customer behaviour.

### Retraining

The model should be periodically retrained using more recent customer data to maintain predictive accuracy.

### Integration

Predicted churn probabilities should be integrated with existing Customer Relationship Management (CRM) systems so that customer service teams can prioritise high-risk customers for intervention.

## 5. Project Limitations

Several limitations should be considered when interpreting the results.

- The analysis relied on a single publicly available dataset.
- Customer behaviour may evolve over time, requiring periodic model retraining.
- The model identifies statistical associations rather than causal relationships.
- External factors such as macroeconomic conditions and competitor activity were unavailable.
- Customer feedback and satisfaction metrics were not incorporated into the predictive models.

## 6. Future Improvements

Potential extensions include:

- evaluating additional ensemble algorithms such as XGBoost and LightGBM
- incorporating temporal behavioural features
- implementing cost-sensitive learning
- integrating real-time prediction pipelines
- developing interactive dashboards for business users
- deploying the model using cloud-based machine learning services
